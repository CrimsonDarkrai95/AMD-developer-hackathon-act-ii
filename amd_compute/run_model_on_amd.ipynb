# AMD Compute — Run a Model on AMD Hardware (Track 3 requirement)

This notebook does exactly one thing: pull and run an LLM on this AMD instance via
Ollama, and log proof it ran. Nothing else in this project depends on this notebook —
the live backend/frontend (GitHub repo) continues to use the Featherless API and is
untouched by anything here.

**Upload just this notebook file to the AMD Jupyter environment and run it top to
bottom. Do not clear the outputs before committing the .ipynb back to the repo** —
the cell outputs are the proof of AMD compute usage.

## 1. Confirm this is AMD hardware

```python
import subprocess
print(subprocess.run(['rocm-smi'], capture_output=True, text=True, timeout=10).stdout)
```

## 2. Start Ollama and pull a model

Swap `MODEL_NAME` for whichever model you want to showcase.

```python
import subprocess, time

MODEL_NAME = 'llama3'  # change to whatever model you want to run

# Start the ollama server in the background if it isn't already running
subprocess.Popen(['ollama', 'serve'])
time.sleep(5)

pull = subprocess.run(['ollama', 'pull', MODEL_NAME], capture_output=True, text=True)
print(pull.stdout)
print(pull.stderr)
```

## 3. Run the model and capture output as proof

```python
import json
from pathlib import Path

PROMPT = 'In one sentence, explain what a nephropathy risk screen looks for.'

run = subprocess.run(['ollama', 'run', MODEL_NAME, PROMPT], capture_output=True, text=True)
print(run.stdout)

OUT_DIR = Path('outputs')
OUT_DIR.mkdir(exist_ok=True)
with open(OUT_DIR / 'model_run_log.txt', 'w') as f:
    f.write(f'Model: {MODEL_NAME}\n')
    f.write(f'Prompt: {PROMPT}\n')
    f.write(f'Output:\n{run.stdout}\n')
print('Saved outputs/model_run_log.txt')
```
